```python
# Activity 14
# Importing Libraries

import pandas as pd
from dateutil.parser import parse
```

```python
# Define the path for the dataset and read the data:

filepath = '/content/drive/MyDrive/Dataset/AirQualityUCI.csv'
```

```python
# Read the data

data = pd.read_csv(filepath, delimiter = ";")
data.head()
```

```python
# Drop the unwanted columns:

data = data.drop(['Unnamed: 15', 'Unnamed: 16'], axis = 1)
data.head()
```

```python
# Data Shape

data.shape
```

```python
# Remove all NA values:

data = data.dropna()
data.shape
```

```python
# Check whether there are any data points with NA:

data.info()
```

```python
# Parse the date column and extract the weekday, day,
# and month, as follows:

# Parsing the date
data['Parse_date'] = data['Date'].apply(lambda x: parse(x))
# Parsing the weekday
data['Weekday'] = data['Parse_date'].apply(lambda x: x.weekday())
# Parsing the Day
data['Day'] = data['Parse_date'].apply(lambda x: x.strftime("%A"))
# Parsing the Month
data['Month'] = data['Parse_date'].apply(lambda x: x.strftime("%B"))
data.head()
```

```python
# Question 1: Which day of the week has
# the highest NO2(GT) emissions?

# Aggregate NO2(GT) based on Day and find
# the mean emissions for the days.
Q1_1 = pd.DataFrame(data.groupby(['Day'])['NO2(GT)'].agg('mean'))
Q1_1
```

```python
# Make a bar plot with the data:

DayPlot = Q1_1.plot.bar( y = 'NO2(GT)', rot = 90,\
                        title = 'Emissions on week days')
DayPlot.set_xlabel("Week Days")
DayPlot.set_ylabel("Average Emissions")
```

```python
# Question 2: At what time of the day are NMHC(GT) emissions highest?

Q2_1 = pd.DataFrame(data.groupby(['Time'])['NMHC(GT)'].agg('mean'))
```

```python
# Plot a horizontal bar plot on the data:

timePlot = Q2_1.plot.barh(y = 'NMHC(GT)', \
                          title = 'Emissions at hours of day')
timePlot.set_xlabel("Hours")
timePlot.set_ylabel("Average Emissions")
```

```python
# Question 3: Which month has the highest CO(GT) emissions?

# Create the function to clean
def cleanFeat(x):
  return pd.to_numeric(".".join(x.split(',')))
```

```python
# Cleaning up the CO format
data['CO(GT)'] = data['CO(GT)'].apply(lambda x: cleanFeat(x))
data.head()
```

```python
# Aggregate the data based on the Month
# and find the mean of 'CO(GT)' column

Q3_1 = pd.DataFrame(data.groupby(['Month'])['CO(GT)'].agg('mean'))
Q3_1
```

```python
# Plot the data using a horizonatl bar chart:

monthPlot = Q3_1.plot.barh(y = 'CO(GT)',\
                           title = 'Monthly average emissions')
monthPlot.set_xlabel("Months")
monthPlot.set_ylabel("Average Emissions")
```
