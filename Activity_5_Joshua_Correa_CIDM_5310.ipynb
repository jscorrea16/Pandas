```python
import pandas as pd
```

```python
# Read in the mushroom.csv data

mushroom = pd.read_csv('/content/drive/MyDrive/Dataset/mushroom.csv')

mushroom.head(5)
```

```python
# List the columns

mushroom.columns
```

```python
mushroom_index = pd.MultiIndex.from_frame(mushroom[['class',
                                              'population',
                                              'habitat']])
```

```python
# Drop the columns that are in the index and set the
# DataFrame index to the multi-index.

mushroom.drop(columns = ['class','population','habitat'],inplace = True)

mushroom.set_index(mushroom_index, inplace = True)
mushroom.head(10)
```
