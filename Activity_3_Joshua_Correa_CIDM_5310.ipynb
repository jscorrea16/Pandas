```python
# Libraries

import pandas as pd
import sqlite3
```

```python
# Tables in Supply_Company Database


tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type = 'table'",
                     sqlite3.connect('/content/drive/MyDrive/Dataset/supply_company.db'))
tables
```

```python
# Customer Table in the Supply_Company Database


customers = pd.read_sql("SELECT * FROM Customers",
                        sqlite3.connect('/content/drive/MyDrive/Dataset/supply_company.db'))
customers
```

```python
# Orders Table in the Supply_Company Database

orders = pd.read_sql("SELECT * FROM Orders",
                     sqlite3.connect('/content/drive/MyDrive/Dataset/supply_company.db'))
orders
```

```python
# Largest Customer Sales

largest_cust = \
  orders.groupby('Customer_Number').sum()['amount'].\
                  sort_values(ascending = False).index[0]
largest_cust
```

```python
# Largest Customer Information

largest_customer_info =\
  pd.read_sql(("SELECT * FROM Customers WHERE Customer_Number = " +
               str(largest_cust)),
              sqlite3.connect("/content/drive/MyDrive/Dataset/supply_company.db"))

largest_customer_info
```
