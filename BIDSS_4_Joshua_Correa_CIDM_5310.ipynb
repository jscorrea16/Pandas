```python
def highlight_col(x):
  r = 'background-color: red'
  g = 'background-color: grey'
  temp_df = pd.DataFrame('', index = x.index, columns = x.columns)
  temp_df.iloc[:, 2] = r
  temp_df.iloc[:, 3] = g
  return temp_df

cdc_state_df.style.apply(highlight_col, axis = None)
```

```python
def state_cases_for_state(state):
  labels = ['State', 'Reported Cases']
  colors = ['blue', 'red']
  mode_size = [6 , 8]
  line_size = [4 , 5]

  df_list = [cdc_state_df, cdc_monthly_illness_df]

  fig = go.Figure()

  for i, df in enumerate(df_list):
    if state == "State" or state == 'state':
      x_data = np.array(list(df.iloc[:, 5].columns))
      y_data = np.sum(np.asarray(df.iloc[:, 5]), axis = 0)

    else:
            x_data = np.array(list(df.iloc[:, 5:].columns))
            y_data = np.sum(np.asarray(df[df['State'] == state].iloc[:,5:]),axis = 0)

            fig.add_trace(go.Scatter(x=x_data, y=y_data, mode = 'lines+markers',
                          name = labels[i],
                          line = dict(color =colors[i], width = line_size[i]),
                          connectgaps = True,
                          text = "Total " + str(labels[i]) + ": "+ str(y_data[-1])

                          ))
  fig.show()



interact(state_cases_for_state, state = 'State');
```
