```python
#  Exercise 3.02 - Text Character Encoding and Data Separators
# sqlite3
# See VSCode Folder

import pandas as pd
import sqlite3
conn = sqlite3.connect('Dataset/bike_share.db')
c = conn.cursor()
data = pd.read_csv('Dataset/bike_share_UCS_2_LE_BOM.tsv',
                  encoding = 'utf_16_le',
                  sep = '\t')
c.execute('CREATE TABLE IF NOT EXISTS RENTALS(Date,Hour,Qty)')
conn.commit()
data.to_sql("RENTALS",conn,if_exists='replace')
```
